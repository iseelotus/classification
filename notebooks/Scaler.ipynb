```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
matplotlib.style.use('ggplot')
```

```python
# Standard Scaler

np.random.seed(42)
df = pd.DataFrame({
    'x1': np.random.normal(0, 5, 5000),
    'x2': np.random.normal(10, 15, 5000),
    'x3': np.random.normal(-10, 7, 5000)
})

scaler = preprocessing.MinMaxScaler()
scaled_df = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_df, columns=['x1', 'x2', 'x3'])

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(6, 5))

ax1.set_title('Before Scaling')
sns.kdeplot(df['x1'], ax=ax1)
sns.kdeplot(df['x2'], ax=ax1)
sns.kdeplot(df['x3'], ax=ax1)
ax2.set_title('After Standard Scaler')
sns.kdeplot(scaled_df['x1'], ax=ax2)
sns.kdeplot(scaled_df['x2'], ax=ax2)
sns.kdeplot(scaled_df['x3'], ax=ax2)
plt.show()
```

```python
# min-max scaler
df = pd.DataFrame({
    # positive skew
    'x1': np.random.chisquare(4, 500),
    # negative skew 
    'x2': np.random.beta(8, 2, 500) * 40,
    # no skew
    'x3': np.random.normal(50, 3, 500)
})

scaler = preprocessing.MinMaxScaler()
minmax_scaled_df = scaler.fit_transform(df)
minmax_scaled_df = pd.DataFrame(minmax_scaled_df, columns=['x1', 'x2', 'x3'])

scaler = preprocessing.RobustScaler()
robust_scaled_df = scaler.fit_transform(df)
robust_scaled_df = pd.DataFrame(robust_scaled_df, columns=['x1', 'x2', 'x3'])

fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(9, 5))
ax1.set_title('Before Scaling')
sns.kdeplot(df['x1'], ax=ax1)
sns.kdeplot(df['x2'], ax=ax1)
sns.kdeplot(df['x3'], ax=ax1)
ax2.set_title('After Min-Max Scaling')
sns.kdeplot(minmax_scaled_df['x1'], ax=ax2)
sns.kdeplot(minmax_scaled_df['x2'], ax=ax2)
sns.kdeplot(minmax_scaled_df['x3'], ax=ax2)
ax3.set_title('After Robust Scaling')
sns.kdeplot(robust_scaled_df['x1'], ax=ax3)
sns.kdeplot(robust_scaled_df['x2'], ax=ax3)
sns.kdeplot(robust_scaled_df['x3'], ax=ax3)
plt.show()
```
